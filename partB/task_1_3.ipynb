# Task 1.3 — Claims to Improve: MKL vs. Single-Kernel SVM

**Paper**: Gönen, M. & Alpaydın, E. (2011). *Multiple Kernel Learning Algorithms*. JMLR, 12, 2211–2268.

---

## Baseline: Single-Kernel SVM

The baseline is a standard Support Vector Machine (SVM) trained with a **single, fixed kernel function** — for instance, a linear kernel or an RBF kernel with a particular bandwidth $\gamma$. This is the most common approach prior to MKL.

## Limitation of the Baseline

A single-kernel SVM requires the practitioner to **manually select** the kernel type and its hyperparameters. This is problematic because:

1. **No universal kernel**: No single kernel is optimal for all datasets. A linear kernel fails on non-linear boundaries; an RBF kernel with a poorly chosen $\gamma$ either underfits (too large) or overfits (too small).
2. **Single data view**: A fixed kernel imposes a single similarity measure, which cannot simultaneously capture global linear structure and local non-linear clusters.
3. **Expensive search**: Grid search over kernel types and hyperparameters is combinatorially expensive and does not scale gracefully to heterogeneous or multi-source data.

## How MKL Overcomes This Limitation

MKL replaces the manual kernel selection step with an **automatic, data-driven kernel combination**:

- It defines a set of candidate kernels (different types and/or hyperparameter settings).
- It learns non-negative weights $\boldsymbol{\eta} = (\eta_1, \dots, \eta_P)$ that combine these kernels into a single, task-optimal kernel.
- The kernel weights and SVM parameters are optimised jointly (or alternately), so the model automatically favours kernels that improve classification and down-weights irrelevant ones.

**Result**: MKL matches or exceeds the performance of the best single kernel without requiring exhaustive kernel selection, and provides interpretable kernel weights as a by-product (Sections 6–7).

## Failure Condition: When MKL Would NOT Outperform the Baseline

MKL is unlikely to outperform a well-tuned single-kernel SVM when:

> **The optimal decision boundary is fully captured by a single kernel, and all additional kernels are redundant or noisy.**

For example, if the data is perfectly linearly separable and the candidate kernel set consists of a linear kernel plus several RBF kernels, the ideal solution assigns $\eta_{\text{linear}}=1$ and $\eta_{\text{RBF}}=0$ — recovering exactly the single-kernel baseline. In practice, the optimisation overhead and potential numerical issues in MKL may slightly *hurt* performance compared to simply using the linear kernel directly. The $\ell_1$ constraint may also introduce instability when all kernels are near-equivalent, as small perturbations in the data can change which kernel(s) receive non-zero weight (Section 7.2).
