# Task 1.2 — Key Assumptions of Multiple Kernel Learning

**Paper**: Gönen, M. & Alpaydın, E. (2011). *Multiple Kernel Learning Algorithms*. JMLR, 12, 2211–2268.

---

## Assumption 1: Base Kernels Capture Complementary Information

**Assumption:** Each base kernel encodes a distinct and complementary notion of similarity. For example, a linear kernel captures global linear trends while an RBF kernel captures local non-linear structure.

**Why the method needs it:** If kernels were perfectly redundant, combining them would add model complexity without any representational gain; the combined kernel would offer no advantage over the best single kernel.

**Violation scenario:** If all base kernels are nearly identical (e.g., multiple RBF kernels with very similar bandwidths on the same features), MKL degenerates into a re-weighted single kernel and provides no benefit over kernel selection.

**Paper reference:** Section 1 (Introduction), Section 7 (Discussion): *"The premise of MKL is that using multiple kernels… allows integrating different data sources or representations."*

---

## Assumption 2: Data Is Separable in the Combined Kernel Feature Space

**Assumption:** There exists some combination of the base kernels such that the data becomes (approximately) linearly separable in the resulting combined Reproducing Kernel Hilbert Space (RKHS).

**Why the method needs it:** SVM-based MKL still relies on finding a maximum-margin hyperplane in the combined feature space. If no combination of kernels produces a space where the classes are separable, the margin will be small and generalisation will be poor.

**Violation scenario:** When features are pure noise and completely unrelated to the target label, no kernel combination can create meaningful class separation. The SVM will overfit to noise, yielding near-chance accuracy.

**Paper reference:** Section 2 (Eq. 5–7): the combined kernel defines the RKHS in which the SVM operates; Section 4 (optimisation assumes a feasible solution with positive margin).

---

## Assumption 3: Kernel Weight Sparsity Leads to Better Generalisation

**Assumption:** Encouraging sparse kernel weights (e.g., via an $\ell_1$-norm constraint $\sum_m \eta_m = 1$, $\eta_m \geq 0$) acts as an effective regulariser, selecting only the most task-relevant kernels and discarding noisy or irrelevant ones.

**Why the method needs it:** Without sparsity, the model may assign non-trivial weight to irrelevant kernels, diluting the discriminative signal and harming generalisation — analogous to including irrelevant features in a model.

**Violation scenario:** When *all* base kernels carry equally important and non-redundant information, enforcing sparsity is harmful because it suppresses useful kernels. In this case, a uniform (dense) combination is superior.

**Paper reference:** Section 3.2 & Section 4 (SimpleMKL uses $\ell_1$ constraint on $\boldsymbol{\eta}$); Section 7.2 discusses the sparsity vs. non-sparsity trade-off.
